```python
import pandas as pd

import numpy as np

usefile = pd.read_excel('repayment_base_file.xlsx')

TM = pd.read_excel('TM.xlsx')
```

```python
sps = pd.read_excel('BI NG daily reports of 2023-08-30.xlsx','Sps Contracts', usecols = ['Contract','activation_end'])

last_pay = pd.read_excel('BI NG daily reports of 2023-08-30.xlsx','Last payments', usecols = ['Contract','activation_end'])
```

```python
import datetime as date
from datetime import date
from datetime import timedelta

last_pay['activation_end'] = pd.to_datetime(last_pay['activation_end'], errors = 'coerce').dt.normalize()
sps['activation_end'] = pd.to_datetime(sps['activation_end']).dt.normalize()

sps.columns = ['Contract','sps_activation_end']

last_pay.columns = ['Contract','payment_activation_end']
```

```python
merge1 = pd.merge(pd.merge(usefile, sps, how='left', on='Contract'), last_pay, how='left', on='Contract')
```

```python
merge1
```

```python
merge1["due_date"] = merge1[["sps_activation_end","payment_activation_end"]].max(axis=1)
```

```python
merge1 = merge1[['Contract','phone_number','LCP','customer_name','contract_number','tm','due_date','Sold_By_Exited_TM','Sold_By_Exited_LCP']]
```

```python
merge1 = pd.merge(merge1, TM, left_on='contract_number', right_on='Contract', how='left')
```

```python
merge1['New TM'].fillna(merge1['tm'], inplace=True)
```

```python
merge1['New LCP'].fillna(merge1['LCP'], inplace=True)
```

```python
merge1
```

```python
daycard = pd.to_datetime('today').normalize() - timedelta(days=1)

merge1['days'] = (daycard - merge1['due_date']).dt.days
```

```python
merge1 = merge1[(merge1['days'] >= -5) & (merge1['days'] != 0)]
```

```python
filters = [
   (merge1.days < 0),
   (merge1.days <= 5),
   (merge1.days <= 30)
]
values = ['Upcoming_Payments', 'Late_Payments', 'Retrieval']

merge1['repayment_status'] = np.select(filters, values, default= 'Penalty')
```

```python
output = merge1.groupby(['New TM','New LCP','Sold_By_Exited_TM','Sold_By_Exited_LCP','contract_number','customer_name','phone_number','due_date','repayment_status'])['days'].max().unstack()
```
