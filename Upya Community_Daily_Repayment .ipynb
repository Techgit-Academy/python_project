### Upya Community_Daily_Repayment

```python
import pandas as pd
import numpy as np

repayment_base = pd.read_excel('Upya_repayment_base_file.xlsx')
```

```python
payment = pd.read_csv('contracts upya 2023-08-30b.csv')
```

```python
repayment_base = pd.merge(repayment_base, payment, how='left', on='contractNumber')
```

```python
repayment_base = repayment_base[['client_number','contractNumber','State','customer_name','TM','LCP','totalDaysActivated','phone_number','signingDate']]
```

```python
repayment_base
```

```python
import datetime as date
from datetime import date
from datetime import timedelta
import numpy as np

daycard = pd.to_datetime('today').normalize()
daycard
```

```python
repayment_base['signingDate'] = pd.to_datetime(repayment_base['signingDate'].str[0:10]).dt.normalize()
#repayment_base['signingDate'] = pd.to_datetime(repayment_base['signingDate'], format='%Y/%m/%d')
```

```python
repayment_base["age"] = (daycard - repayment_base['signingDate']).dt.days
```

```python
repayment_base["days_deficit"] = repayment_base['age'] - repayment_base['totalDaysActivated']
```

```python
#repayment_base["Due_Date"] = repayment_base['signingDate'] + repayment_base['totalDaysActivated'].dt.days

repayment_base["Due_Date"] = repayment_base['signingDate']+pd.to_timedelta(repayment_base['totalDaysActivated'],unit='d')
```

```python
repayment_base.head(10)
```

```python
repayment_base
```

```python
repayment_base = repayment_base[(repayment_base['days_deficit'] >= -5) & (repayment_base['days_deficit'] != 0)]
```

```python
filters = [
   (repayment_base.days_deficit < 0),
   (repayment_base.days_deficit <= 5),
   (repayment_base.days_deficit <= 30)
]
values = ['Upcoming_Payments', 'Late_Payments', 'Retrieval']

repayment_base['repayment_status'] = np.select(filters, values, default= 'Penalty')
```

```python
repayment_base = repayment_base.groupby(['client_number','TM','LCP','contractNumber','customer_name','phone_number','Due_Date','repayment_status'])['days_deficit'].max().unstack()
```
